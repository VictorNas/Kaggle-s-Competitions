```python
Y = my_train[['revenue']]
```

### Separando a base de dados em Treino e Validação

```python
from sklearn.model_selection import train_test_split
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(X, Y, test_size = 0.2)
```

```python
print("X_treinamento"+ str(X_treinamento.shape))
print('\n')
print("y_treinamento"+ str(y_treinamento.shape))
print('\n')
print("X_teste"+ str(X_teste.shape))
print('\n')
print("y_teste"+ str(y_teste.shape))
print('\n')
```

## Criando Rede Neural com TensorFlow

```python
X_treinamento
```

```python
import tensorflow as tf
from keras.layers import Input, Dense
from keras.models import Model
from keras.callbacks import EarlyStopping
import keras.optimizers
tf.compat.v1.get_default_graph()
```

```python
inputs = Input(shape=(80,))

camada1 = Dense(80, activation='relu')(inputs)
camada2 = Dense(64, activation='relu')(camada1)
camada3 = Dense(32, activation='relu')(camada2)
camada4 = Dense(16, activation='relu')(camada3)
camada5 = Dense(8, activation='relu')(camada4)
predictions = Dense(1, activation='relu')(camada5)

optim = keras.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
model = Model(inputs=inputs, outputs=predictions)
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=20)
```

```python
type(es)
```

```python
model.compile(optimizer=optim,
              loss='mean_squared_logarithmic_error',
              metrics=['mean_squared_error'])
model.fit(X_treinamento, y_treinamento, epochs = 1000,batch_size= 32,validation_data=(X_teste,y_teste),callbacks=[es])  # starts training
```

```python
previsoes = model.predict(X_teste)
```

```python
y_teste
```

```python
from sklearn.metrics import mean_squared_log_error
from math import sqrt
sqrt(mean_squared_log_error(y_teste,previsoes))
```
